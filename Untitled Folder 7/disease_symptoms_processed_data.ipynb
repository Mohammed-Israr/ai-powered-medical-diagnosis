```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Load dataset (replace with your actual data)
df = pd.read_csv('disease_symptoms.csv')

# Fill null values with 0
df.fillna(0, inplace=True)
df

from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

symptom_columns = df.columns[1:]  
df[symptom_columns] = df[symptom_columns].astype(str)

encoded_symptoms = ohe.fit_transform(df[symptom_columns])

encoded_df = pd.DataFrame(encoded_symptoms, columns=ohe.get_feature_names_out(symptom_columns))

final_df = pd.concat([df[['Disease']], encoded_df], axis=1)
final_df
```
